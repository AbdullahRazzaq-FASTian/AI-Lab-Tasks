# **Task 1**

```python
class Environment:
    def __init__(self):
        self.beds = ['Moist'] * 9
        self.beds[1] = 'Dry'
        self.beds[4] = 'Dry'
        self.beds[6] = 'Dry'

    def get_percept(self, position):
        return self.beds[position]

    def water_bed(self, position):
        self.beds[position] = 'Moist'

    def display_beds(self):
        print(f"Final Bed Status: {self.beds}")

class SimpleReflexAgent:
    def __init__(self):
        self.position = 0

    def act(self, percept):
        if percept == 'Dry':
            return 'Water the bed'
        else:
            return 'Move to next'

    def move(self):
        if self.position < 8:
            self.position += 1

def run_agent(agent, environment):
    print("Starting Irrigation Process...")
    for _ in range(9):
        percept = environment.get_percept(agent.position)
        action = agent.act(percept)

        print(f"Bed {agent.position + 1}: {percept} -> {action}")

        if action == 'Water the bed':
            environment.water_bed(agent.position)
            print(f"   -> Bed {agent.position + 1} is now Moist")

        agent.move()

    environment.display_beds()

agent = SimpleReflexAgent()
environment = Environment()
run_agent(agent, environment)
```

# **Task 2**

```python
class Environment:
    def __init__(self):
        self.key_location = 'B'
        self.vault_location = 'D'

    def get_percept(self, location):
        return location

class SecurityRobot:
    def __init__(self):
        self.has_key = False

    def act(self, location):
        if location == 'B':
            self.has_key = True
            return "Picked up keycard"
        elif location == 'D':
            if self.has_key:
                return "Access Granted to Vault"
            else:
                return "Access Denied - No Keycard"
        else:
            return "Patrolling"

def run_simulation():
    env = Environment()
    robot = SecurityRobot()
    path = ['A', 'C', 'D', 'B', 'D']

    print("Starting Patrol...")
    for room in path:
        action = robot.act(room)
        print(f"Visiting Room {room}: {action}")

run_simulation()
```

# **Task 3**

```python
class Environment:
    def __init__(self):
        self.houses = ['Blue House', 'Green House', 'Red House', 'Yellow House', 'White House']

    def get_percept(self, index):
        return self.houses[index]

class DeliveryDrone:
    def __init__(self):
        self.goal = 'Red House'

    def act(self, percept):
        if percept == self.goal:
            return 'Deliver Package'
        return 'Fly to Next House'

def run_mission():
    environment = Environment()
    drone = DeliveryDrone()

    for i in range(len(environment.houses)):
        current_house = environment.get_percept(i)
        action = drone.act(current_house)

        print(f"Scanning {current_house}: {action}")

        if action == 'Deliver Package':
            print("Success: Package Delivered to Red House. Returning to Base.")
            break

run_mission()
```

# **Task 4**

```python
class Environment:
    def __init__(self):
        self.rocks = [
            {'name': 'Rock A', 'value': 5, 'cost': 2},
            {'name': 'Rock B', 'value': 9, 'cost': 8},
            {'name': 'Rock C', 'value': 6, 'cost': 3}
        ]

    def get_percept(self):
        return self.rocks

class UtilityBasedAgent:
    def __init__(self):
        self.best_rock = None
        self.highest_utility = -1

    def utility_function(self, rock):
        return (rock['value'] * 2) - rock['cost']

    def act(self, rocks):
        for rock in rocks:
            score = self.utility_function(rock)
            print(f"Analyzing {rock['name']}: Utility {score}")

            if score > self.highest_utility:
                self.highest_utility = score
                self.best_rock = rock['name']

        return self.best_rock

def run_mission():
    environment = Environment()
    rover = UtilityBasedAgent()

    rocks = environment.get_percept()
    decision = rover.act(rocks)

    print(f"Mission Target Selected: {decision}")

run_mission()
```

# **Task 5**

```python
import random

class Environment:
    def __init__(self):
        # 0: Start, 1: Empty, 2: Cheese, 3: Trap
        self.track = ['Start', 'Empty', 'Cheese', 'Trap']
        self.rewards = {'Start': 0, 'Empty': -1, 'Cheese': 10, 'Trap': -10}

    def get_reward(self, position):
        state = self.track[position]
        return state, self.rewards[state]

class LearningAgent:
    def __init__(self):
        self.position = 0  # Start at index 0
        self.total_score = 0

    def move(self):
        # Simulate a random move to index 1, 2, or 3
        # (Assuming the mouse leaves the start position)
        self.position = random.randint(1, 3)
        return self.position

def run_simulation():
    env = Environment()
    mouse = LearningAgent()

    print(f"Mouse starts at: {env.track[mouse.position]}")

    # Mouse takes an action (Trial)
    new_pos = mouse.move()
    state, reward = env.get_reward(new_pos)

    mouse.total_score += reward

    # Feedback (Error/Success signal)
    print(f"Mouse moved to: {state}")
    print(f"Reward received: {reward}")
    print(f"Total Score: {mouse.total_score}")

if __name__ == "__main__":
    run_simulation()
```
