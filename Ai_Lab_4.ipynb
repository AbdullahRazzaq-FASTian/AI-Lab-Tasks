```python
from collections import deque

graph = {
    'Tehran' : ['Baghdad', 'Istanbul'],
    'Baghdad' : ['Cairo'],
    'Istanbul' : ['Berlin'],
    'Cairo' : ['Washington'],
    'Berlin' : ['Washington']
}

def bfs(graph, start, goal):
  visited = []
  dq = deque([])
  dq.append(start)
  visited.append(start)
  while dq:
    cur = dq.popleft();
    if(cur == goal):
      return visited
    for neighbour in graph.get(cur, []):
      if neighbour not in visited:
        visited.append(neighbour)
        dq.append(neighbour)
  return []

visited = bfs(graph, 'Tehran', 'Washington')
if(len(visited) == 0):
  print("Goal not found")
else:
  print("Goal found")
  print(f"Visits: {" -> ".join(map(str, visited))}")
```

```python
import heapq

graph = {
    'Earth': {'Moon_Base': 10, 'Orbital_Platform': 5},
    'Orbital_Platform': {'Moon_Base': 2, 'Mars': 60},
    'Moon_Base': {'Mars': 50},
    'Mars': {}
}

def ucs(graph, start, goal):
    visited = set()
    pq = []
    heapq.heappush(pq, (0, start, [start]))

    while pq:
        current_cost, cur, path = heapq.heappop(pq)

        if cur == goal:
            return path, current_cost

        if cur not in visited:
            visited.add(cur)

            for neighbour, weight in graph.get(cur, {}).items():
                if neighbour not in visited:
                    new_cost = current_cost + weight
                    new_path = path + [neighbour]
                    heapq.heappush(pq, (new_cost, neighbour, new_path))

    return [], 0

path, total_cost = ucs(graph, 'Earth', 'Mars')

if len(path) == 0:
    print("Goal not found")
else:
    print("Goal found")
    print(f"Path: {' -> '.join(path)}")
    print(f"Total Fuel Cost: {total_cost}")
```

```python
graph = {
    'Entrance': ['Hallway_A', 'Hallway_B'],
    'Hallway_A': ['Storage'],
    'Hallway_B': ['Target'],
    'Storage': ['Deep_Vault'],
    'Target': [],
    'Deep_Vault': []
}

def dls(graph, start, goal, limit):
    visited = []

    def dfs(node, depth):
        if depth > limit:
            return None

        visited.append(node)

        if node == goal:
            return visited

        for neighbor in graph.get(node, []):
            if neighbor not in visited:
                path = dfs(neighbor, depth + 1)
                if path:
                    return path

        visited.pop()
        return None

    return dfs(start, 0)

path = dls(graph, 'Entrance', 'Target', 2)

if path:
    print("Mission Accomplished!")
    print(f"Path: {' -> '.join(path)}")
else:
    print("Mission Failed: Target not found within depth limit")
```

```python
graph = {
    'Main_Box': ['Zone_A', 'Zone_B'],
    'Zone_A': ['Switch_1', 'Switch_2'],
    'Zone_B': ['Breaker_101'],
    'Switch_1': [],
    'Switch_2': [],
    'Breaker_101': []
}

def dls(graph, start, goal, limit):
    visited = []

    def dfs(node, depth):
        if depth > limit:
            return None

        visited.append(node)

        if node == goal:
            return visited

        for neighbor in graph.get(node, []):
            if neighbor not in visited:
                path = dfs(neighbor, depth + 1)
                if path:
                    return path

        visited.pop()
        return None

    return dfs(start, 0)

def ids(graph, start, goal, max_depth):
    for depth in range(max_depth + 1):
        print(f"Checking Depth {depth}...")
        path = dls(graph, start, goal, depth)
        if path:
            return path
    return None

path = ids(graph, 'Main_Box', 'Breaker_101', 2)

if path:
    print("Lights are back on!")
    print(f"Path: {' -> '.join(path)}")
else:
    print("Failed to restore power.")
```
